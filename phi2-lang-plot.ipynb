````python
# --- 1. Installations ---
# For now use these versions as they work together. 
!pip install -q torch==2.3.1 torchvision==0.18.1 torchaudio==2.3.1 --index-url https://download.pytorch.org/whl/cu121
!pip install -q transformers==4.41.2 accelerate==0.30.1 peft==0.11.1 trl==0.8.6 bitsandbytes==0.43.1
!pip install -q sentence-transformers datasets einops pandas matplotlib scipy tqdm
!pip install -q fsspec==2024.10.0


# Install IPython for display if not already present
try:
    from IPython.display import display, Image, clear_output
except ImportError:
    print("Installing IPython for image display...")
    !pip install -q IPython
    from IPython.display import display, Image, clear_output

# --- 2. Imports ---
import pandas as pd
import numpy as np
import re
import torch
import time
import json
import os
import gc
import inspect                    
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments,
    pipeline
)
from peft import (
    LoraConfig,
    PeftModel,
    get_peft_model,
    prepare_model_for_kbit_training
)
from trl import SFTTrainer
from datasets import Dataset
import matplotlib.pyplot as plt
from scipy import stats
from IPython.display import display, Image # Use specific imports
from dataclasses import dataclass, field
from enum import Enum
from typing import Dict, Callable, Any, List, TypedDict, Sequence, Optional
import operator



# --- 3. Configuration ---
MODEL_NAME = "microsoft/phi-2" # this version can be replaced with others.
OUTPUT_DIR = "/kaggle/working/phi2_structured_router_finetune_v2" #  output dir
DATA_CSV_PATH = "/kaggle/working/data.csv"
TEMP_DATA_PATH = '/kaggle/input/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCity.csv'

# --- 4. Setup Directories and Dummy Data ---
if not os.path.exists('/kaggle/working/'): os.makedirs('/kaggle/working/')
if not os.path.exists('/kaggle/input/'): # Dummy data setup
    os.makedirs('/kaggle/input/climate-change-earth-surface-temperature-data/', exist_ok=True)
    if not os.path.exists(TEMP_DATA_PATH): print("Creating dummy data..."); dummy_data={'dt': pd.to_datetime(['1900-01-01','1950-01-01','2000-01-01','2010-01-01']*8),'AverageTemperature': [5,7,9,10]*8,'City': ['Berlin']*4+['Abiko']*4+['Munich']*4+['Tokyo']*4+['Hamburg']*4+['Paris']*4+['London']*4+['New York']*4,'Country': ['Germany']*4+['Japan']*4+['Germany']*4+['Japan']*4+['Germany']*4+['France']*4+['UK']*4+['USA']*4}; pd.DataFrame(dummy_data).to_csv(TEMP_DATA_PATH, index=False); print(f"Dummy data created at {TEMP_DATA_PATH}")

# --- 5. Initial Data Setup Function ---
def load_and_filter_data_for_setup(csv_path, countries, year_start, year_end):
    """Loads, filters, and saves the climate data ONCE to DATA_CSV_PATH."""
    effective_path = csv_path
    print(f"Attempting initial load & filter from: {effective_path} to create {DATA_CSV_PATH}")
    if os.path.exists(DATA_CSV_PATH):
        print(f"{DATA_CSV_PATH} already exists. Skipping initial creation.")
        try: pd.read_csv(DATA_CSV_PATH, nrows=5); print(f"Checked {DATA_CSV_PATH} OK."); return DATA_CSV_PATH
        except Exception as e: print(f"Error reading existing {DATA_CSV_PATH}: {e}. Will attempt recreate.")

    try:
        if not os.path.exists(effective_path) and '/kaggle/input/' in effective_path:
             effective_path = TEMP_DATA_PATH
             if not os.path.exists(effective_path): raise FileNotFoundError(f"Source/Dummy path not found: {effective_path}")
             print(f"Warning: Using dummy data from '{effective_path}'.")
        df = pd.read_csv(effective_path).dropna(subset=['AverageTemperature'])
        print(f"Loaded {len(df)} records from {effective_path}.")
    except Exception as e: print(f"Error loading source CSV {effective_path}: {e}"); return None

    df_filtered = df[['dt', 'AverageTemperature', 'City', 'Country']].copy()
    df_filtered['dt'] = pd.to_datetime(df_filtered['dt']); df_filtered['year'] = df_filtered['dt'].dt.year
    df_filtered = df_filtered[df_filtered['Country'].isin(countries)]
    df_filtered = df_filtered[(df_filtered['year'] >= year_start) & (df_filtered['year'] <= year_end)]
    print(f"Filtered: {len(df_filtered)} records for {countries} ({year_start}-{year_end}).")
    if df_filtered.empty: print("Error: No data after filtering."); return None
    try:
        df_filtered.to_csv(DATA_CSV_PATH, index=False); print(f"Filtered data saved to {DATA_CSV_PATH}.")
        return DATA_CSV_PATH
    except Exception as e: print(f"Error saving filtered data: {e}"); return None

# --- 6. Type-Safe Core System & Tool Registry  ---
class PlotType(Enum): LINE = "line_plot"; SCATTER = "scatter_with_trend"
@dataclass
class PlotRequest:
    city: str; country: str; start_year: int; end_year: int; plot_type: PlotType; title: Optional[str] = None
    def validate(self):
        if not isinstance(self.start_year, int) or not isinstance(self.end_year, int) or self.start_year >= self.end_year: raise ValueError("Invalid year range")
        if not self.city or not self.country: raise ValueError("City/Country missing")
        if not isinstance(self.plot_type, PlotType): raise ValueError(f"Invalid plot type: {self.plot_type}")
        print("PlotRequest validated.")

class ToolRegistry:
    """Manages available plotting tools and their descriptions for the LLM."""
    def __init__(self):
        self._tools: Dict[str, Callable] = {}
        self._descriptions: List[Dict[str, Any]] = []

    def register(self, name: str, description: str, input_params: Dict[str, str]):
        """Decorator to register a plotting function as a tool."""
        def decorator(func: Callable):
            self._tools[name] = func
            tool_info = {
                "name": name, "description": description,
                "required_parameters_from_query": input_params
            }
            self._descriptions = [d for d in self._descriptions if d['name'] != name]
            self._descriptions.append(tool_info)
            print(f"Tool '{name}' registered with input params: {input_params}.")
            return func
        return decorator

    def get_tool(self, name: str) -> Optional[Callable]: # Return Optional Callable
        tool = self._tools.get(name)
        # Return None if not found, let caller handle it
        if tool is None:
             print(f"Warning: Tool '{name}' not found in registry.")
        return tool

    def get_tool_descriptions_json(self) -> str:
        """Returns a JSON string describing the available tools for the LLM prompt."""
        try: return json.dumps(self._descriptions, indent=2)
        except Exception as e: print(f"Error generating tool descriptions: {e}"); return "[]"

registry = ToolRegistry()

# --- 7. Plotting Tools ---
@registry.register(
    name="line_plot",
    description="Generates a line plot of temperature trends over time for a location.",
    input_params={
        "city": "string (e.g., 'Berlin')", "country": "string (e.g., 'Germany')",
        "start_year": "integer (e.g., 1990)", "end_year": "integer (e.g., 2010)",
        "title": "string (optional, e.g., 'Berlin Temps 90-10')" }
)
def plot_line(data: pd.Series, title: str) -> Optional[plt.Figure]: # Return Optional Figure
    if not isinstance(data, pd.Series) or len(data) < 2: print("plot_line: Not enough data."); return None
    try: fig, ax = plt.subplots(figsize=(10, 5)); ax.plot(data.index, data.values, marker='o'); ax.set_title(title); ax.set_xlabel("Year"); ax.set_ylabel("Avg Temp (Â°C)"); ax.grid(True); plt.tight_layout(); print(f"Line plot figure created."); return fig
    except Exception as e: print(f"Error plot_line: {e}"); plt.close(); return None

@registry.register(
    name="scatter_with_trend",
    description="Generates a scatter plot with a linear regression trend line for temperature.",
    input_params={
        "city": "string (e.g., 'Tokyo')", "country": "string (e.g., 'Japan')",
        "start_year": "integer (e.g., 1970)", "end_year": "integer (e.g., 2010)",
        "title": "string (optional, e.g., 'Tokyo Trend Analysis')" }
)
def plot_scatter(data: pd.Series, title: str) -> Optional[plt.Figure]: # Return Optional Figure
    if not isinstance(data, pd.Series) or len(data) < 2: print("plot_scatter: Not enough data."); return None
    try:
        years=data.index.values; temps=data.values;
        if np.all(years == years[0]): print("plot_scatter: Single year value."); return None
        slope, intercept, r_value, _, _ = stats.linregress(years, temps);
        fig, ax = plt.subplots(figsize=(10, 5)); ax.scatter(years, temps, label='Data'); ax.plot(years, intercept + slope*years, 'r', label=f'Trend (RÂ²={r_value**2:.2f})'); ax.set_title(title); ax.set_xlabel("Year"); ax.set_ylabel("Avg Temp (Â°C)"); ax.legend(); ax.grid(True); plt.tight_layout(); print(f"Scatter plot figure created."); return fig
    except Exception as e: print(f"Error plot_scatter: {e}"); plt.close(); return None

# --- 8. Data Pipeline Class ---
class DataPipeline:
    def __init__(self, data_path: str): self.data_path = data_path; self.df = self._load_data()
    def _load_data(self) -> pd.DataFrame:
        if not os.path.exists(self.data_path): raise FileNotFoundError(f"Data not found: {self.data_path}")
        try: df = pd.read_csv(self.data_path); required=['dt','AverageTemperature','City','Country']; assert all(c in df.columns for c in required); df['dt'] = pd.to_datetime(df['dt']); df['year'] = df['dt'].dt.year; print(f"DataPipeline loaded from {self.data_path}"); return df
        except Exception as e: print(f"DataPipeline load error: {e}"); raise
    def filter_data(self, city: str, country: str, start_year: int, end_year: int) -> pd.DataFrame:
        try: mask=(self.df['City'].str.lower()==city.lower()) & (self.df['Country'].str.lower()==country.lower()) & (self.df['year']>=start_year) & (self.df['year']<=end_year); filtered_df = self.df.loc[mask].copy(); print(f"Pipeline: Filtered data - {len(filtered_df)} records."); return filtered_df
        except Exception as e: print(f"Pipeline filter error: {e}"); return pd.DataFrame()
    def calculate_yearly_avg(self, input_df: pd.DataFrame) -> Optional[pd.Series]:
        if not isinstance(input_df, pd.DataFrame) or input_df.empty: return None
        try: yearly_avg = input_df.groupby('year')['AverageTemperature'].mean().dropna(); return yearly_avg
        except Exception as e: print(f"Pipeline average error: {e}"); return None

# --- 9. Training Data Generation (Instruction -> JSON - REVISED) ---
def create_function_call_training_data(num_examples=5000):
    training_samples_list = []
    cities = ['Berlin', 'Tokyo', 'Munich', 'London', 'Paris', 'Cairo', 'Sydney', 'New York', 'Moscow', 'Beijing', 'Rome', 'Madrid']
    countries = ['Germany', 'Japan', 'Germany', 'UK', 'France', 'Egypt', 'Australia', 'USA', 'Russia', 'China', 'Italy', 'Spain']
    plot_type_descs_line = ['line plot', 'line graph', 'temperature progression', 'time series graph', 'yearly trend plot']
    plot_type_descs_scatter = ['scatter plot', 'scatter graph with trend', 'regression analysis plot', 'trend line plot', 'correlation scatter']
    all_plot_type_descs = plot_type_descs_line + plot_type_descs_scatter
    start_years = list(range(1950, 2000, 5)); end_years = list(range(1970, 2014, 5))
    print(f"Generating {num_examples} function call training samples...")
    generated_count = 0; max_attempts = num_examples * 3; attempts = 0
    while generated_count < num_examples and attempts < max_attempts:
        attempts += 1
        try:
            city_idx = np.random.randint(len(cities)); city = cities[city_idx]; country = countries[city_idx]
            plot_type_desc = np.random.choice(all_plot_type_descs); start = np.random.choice(start_years)
            valid_end_years = [y for y in end_years if y > start];
            if not valid_end_years: continue
            end = np.random.choice(valid_end_years); start_int = int(start); end_int = int(end)
            instruction_templates = [f"Gen {plot_type_desc} for {city},{country} {start_int}-{end_int}.", f"Show avg temp {city},{country} {start_int}-{end_int} using {plot_type_desc}.", f"Plot {city} {plot_type_desc} yrs {start_int}-{end_int}."]
            instruction = np.random.choice(instruction_templates)
            if plot_type_desc in plot_type_descs_line: plot_type_value = PlotType.LINE.value
            elif plot_type_desc in plot_type_descs_scatter: plot_type_value = PlotType.SCATTER.value
            else: continue
            output_json = json.dumps({"plot_type": plot_type_value, "parameters": {"city": city, "country": country, "start_year": start_int, "end_year": end_int, "title": f"Avg Temp {city} ({start_int}-{end_int})" }}, indent=2)
            training_samples_list.append({"instruction": instruction.strip(), "output": output_json})
            generated_count += 1
            if generated_count % 500 == 0: print(f"  ...generated {generated_count}")
        except Exception as e: print(f"Warning: Sample gen error: {e}"); continue
    if attempts >= max_attempts: print("Warning: Reached max generation attempts.")
    print(f"Finished generating {len(training_samples_list)} samples.")
    return pd.DataFrame(training_samples_list)

# --- 10. Formatting Function for SFTTrainer ---
tokenizer_for_formatting = None
def get_system_prompt():
    # Ensure registry is populated before getting descriptions
    if not registry._descriptions:
        print("Warning: Tool registry is empty when building system prompt!")
    return f"""You are an expert climate data visualization assistant. Your task is to analyze the user's request and respond ONLY with a single valid JSON object specifying the correct plotting tool and its parameters based on the user's request.

Available Tools:
{registry.get_tool_descriptions_json()}

Respond ONLY with a single valid JSON object containing:
- "plot_type": The 'name' of the tool to use (must be one of {list(registry._tools.keys())}).
- "parameters": A dictionary containing the values extracted from the user request for the fields listed in 'required_parameters_from_query' for the chosen tool. Extract 'city', 'country', 'start_year', and 'end_year'. Infer a suitable 'title' if not specified by the user.

User Request:"""

def format_for_sft(sample):
    """Formats data for fine-tuning: System Prompt + Instruction -> JSON Output."""
    global tokenizer_for_formatting
    if tokenizer_for_formatting is None: raise ValueError("Tokenizer not set")
    instruction = sample['instruction']
    output_json = sample['output']
    system_prompt = get_system_prompt()
    text = f"{system_prompt}\n{instruction}\n\nJSON Output:\n{output_json}{tokenizer_for_formatting.eos_token}"
    return [text]

# --- 11. Fine-tuning Function (QLoRA + SFTTrainer) ---
def finetune_phi2_qlora_structured(training_df):
    """Fine-tunes Phi-2 using QLoRA for structured JSON output."""
    global tokenizer_for_formatting
    print(f"--- Starting QLoRA Fine-tuning for {MODEL_NAME} ---")
    if not isinstance(training_df, pd.DataFrame) or training_df.empty: return None
    adapter_path_result = None; model = None; trainer = None; raw_dataset = None; tokenizer = None

    try:
        tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
        tokenizer_for_formatting = tokenizer
        if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token
        tokenizer.padding_side = 'left'
        tokenizer.model_max_length = 1024
        print(f"Tokenizer loaded. Pad: {tokenizer.pad_token_id}, Side: {tokenizer.padding_side}, MaxLen: {tokenizer.model_max_length}")

        if not all(col in training_df.columns for col in ['instruction', 'output']): raise ValueError("DF needs 'instruction', 'output'")
        training_df['instruction'] = training_df['instruction'].astype(str); training_df['output'] = training_df['output'].astype(str)
        raw_dataset = Dataset.from_pandas(training_df[['instruction', 'output']])
        raw_dataset = raw_dataset.shuffle(seed=42)
        print(f"Dataset loaded: {len(raw_dataset)} samples.")
        print("\nExample formatted prompt for training (first 600 chars):")
        print(format_for_sft(raw_dataset[0])[0][:600] + "...")

        bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16, bnb_4bit_use_double_quant=True)
        print("Loading 4-bit model...")
        model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, quantization_config=bnb_config, device_map="auto", trust_remote_code=True)
        if model.config.pad_token_id is None or model.config.pad_token_id != tokenizer.pad_token_id: model.config.pad_token_id = tokenizer.pad_token_id
        model = prepare_model_for_kbit_training(model, use_gradient_checkpointing=True)
        print("Model loaded & prepared (GC enabled).")

        lora_config = LoraConfig(r=16, lora_alpha=32, target_modules=["Wqkv", "out_proj", "fc1", "fc2"], lora_dropout=0.05, bias="none", task_type="CAUSAL_LM")
        model = get_peft_model(model, lora_config)
        print("LoRA applied:"); model.print_trainable_parameters()

        training_args = TrainingArguments(
            output_dir=OUTPUT_DIR, per_device_train_batch_size=2, gradient_accumulation_steps=4,
            learning_rate=1e-4, num_train_epochs=10,
            logging_dir=f"{OUTPUT_DIR}/logs", logging_steps=20, save_strategy="epoch",
            fp16=True, max_grad_norm=0.3, warmup_ratio=0.1, lr_scheduler_type="cosine", report_to="none",
            dataloader_num_workers=1, save_total_limit=1
        )
        print("TrainingArguments defined.")

        trainer = SFTTrainer(
            model=model, train_dataset=raw_dataset, peft_config=lora_config,
            formatting_func=format_for_sft, tokenizer=tokenizer, args=training_args, packing=False,
        )
        print("SFTTrainer initialized.")

        print("\n--- Starting Training ---"); gc.collect(); torch.cuda.empty_cache()
        train_result = trainer.train(); print("--- Training Finished ---")
        metrics = train_result.metrics; print("\nTraining Metrics:", metrics)
        adapter_path_result = os.path.join(OUTPUT_DIR, "final_adapter")
        print(f"Saving adapter to {adapter_path_result}")
        trainer.save_model(adapter_path_result)
        tokenizer.save_pretrained(adapter_path_result)
        print("Adapter and tokenizer saved.")

    except Exception as e: print(f"Error during fine-tuning: {e}"); import traceback; traceback.print_exc(); adapter_path_result = None
    finally: tokenizer_for_formatting = None; del model, trainer, raw_dataset, tokenizer; gc.collect(); torch.cuda.empty_cache(); print("Cleaned up fine-tuning resources.")
    return adapter_path_result


# --- 12. LLM Function Router Class ---
class LLMFunctionRouter:
    """Orchestrates LLM calls and tool execution using the structured architecture."""
    def __init__(self, model_base_path: str, adapter_path: str, data_pipeline: DataPipeline, registry: ToolRegistry):
        self.model_base_path = model_base_path; self.adapter_path = adapter_path
        self.pipeline = data_pipeline; self.registry = registry
        self.model = None; self.tokenizer = None
        self.SYSTEM_PROMPT = self._build_system_prompt() # Build prompt on init
        print("LLMFunctionRouter initialized. System Prompt built.")
        # Optionally pre-load model here if desired, or keep lazy loading
        # self._load_model_and_tokenizer()

    def _build_system_prompt(self) -> str:
        """Builds the system prompt including dynamic tool descriptions."""
        return f"""You are an expert climate data visualization assistant. Your task is to analyze the user's request and respond ONLY with a single valid JSON object specifying the correct plotting tool and its parameters based on the user's request.

Available Tools:
{self.registry.get_tool_descriptions_json()}

Respond ONLY with a single valid JSON object containing:
- "plot_type": The 'name' of the tool to use (must be one of {list(self.registry._tools.keys())}).
- "parameters": A dictionary containing the values extracted from the user request for the fields listed in 'required_parameters_from_query' for the chosen tool. Extract 'city', 'country', 'start_year', and 'end_year'. Infer a suitable 'title' if not specified by the user.

User Request:"""

    def _load_model_and_tokenizer(self):
        """Loads the fine-tuned QLoRA model and tokenizer if not already loaded."""
        if self.model is not None and self.tokenizer is not None: print("LLM Router: Using pre-loaded model/tokenizer."); return
        print(f"LLM Router: Loading model {self.model_base_path} + adapter {self.adapter_path}..."); gc.collect(); torch.cuda.empty_cache()
        model = None; base_model = None; tokenizer = None # Ensure cleanup scope
        try:
            bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16, bnb_4bit_use_double_quant=True)
            base_model = AutoModelForCausalLM.from_pretrained(self.model_base_path, quantization_config=bnb_config, device_map="auto", trust_remote_code=True)
            model = PeftModel.from_pretrained(base_model, self.adapter_path)
            model = model.merge_and_unload(); model.eval()
            tokenizer = AutoTokenizer.from_pretrained(self.adapter_path, trust_remote_code=True)
            if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token
            tokenizer.padding_side = 'left'
            self.model = model; self.tokenizer = tokenizer # Assign to class attributes
            print("LLM Router: Model and tokenizer loaded successfully.")
        except Exception as e: print(f"LLM Router: Error loading model/tokenizer: {e}"); self.model = None; self.tokenizer = None; raise RuntimeError("LLM Load Failed.") from e
        finally: del base_model # Explicitly delete base model after merge if possible

    def _unload_model(self):
         if self.model is None and self.tokenizer is None: return
         print("LLM Router: Unloading model/tokenizer..."); del self.model; del self.tokenizer; self.model = None; self.tokenizer = None; gc.collect(); torch.cuda.empty_cache(); print("Resources unloaded.")

    def parse_request(self, query: str) -> Optional[PlotRequest]:
        """Generates JSON from LLM and parses it into a PlotRequest."""
        if self.model is None or self.tokenizer is None:
            try: self._load_model_and_tokenizer()
            except RuntimeError: return None

        full_prompt = f"{self.SYSTEM_PROMPT}\n{query}\n\nJSON Output:\n"
        print(f"\nLLM Router: Sending prompt (first 300 chars):\n{full_prompt[:300]}...")
        device = self.model.device
        inputs = self.tokenizer(full_prompt, return_tensors="pt", return_attention_mask=True).to(device)
        generated_text = ""; json_str_cleaned = None; data = None; request = None

        print("LLM Router: Generating JSON...")
        with torch.no_grad():
            try:
                outputs = self.model.generate(
                    input_ids=inputs['input_ids'], attention_mask=inputs['attention_mask'],
                    max_new_tokens=250, do_sample=False,
                    pad_token_id=self.tokenizer.eos_token_id, eos_token_id=self.tokenizer.eos_token_id
                )
                generated_text = self.tokenizer.decode(outputs[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)
                print(f"LLM Router: Raw JSON Output:\n>>>\n{generated_text}\n<<<")
            except Exception as e: print(f"LLM Router: Generation error: {e}"); return None

        try:
            json_str_cleaned = None; generated_text = generated_text.strip()
            json_match_markdown = re.search(r"```(?:json)?\s*(\{.*?\})\s*```", generated_text, re.DOTALL | re.IGNORECASE)
            json_match_braces = None
            if not json_match_markdown: start_brace = generated_text.find('{')
            else: start_brace = -1 # Prevent brace matching if markdown found

            if json_match_markdown: json_str_cleaned = json_match_markdown.group(1).strip(); print("LLM Router: Extracted JSON using markdown markers.")
            elif start_brace != -1:
                brace_level = 0; end_brace = -1
                for i, char in enumerate(generated_text[start_brace:]):
                    if char == '{': brace_level += 1
                    elif char == '}': brace_level -= 1
                    if brace_level == 0: end_brace = start_brace + i; break
                if end_brace != -1: json_str_cleaned = generated_text[start_brace : end_brace + 1].strip(); print("LLM Router: Extracted JSON using brace matching.")
                else: print("Warning: Found opening brace but no matching closing brace.")

            # Fallback only if other methods failed
            if not json_str_cleaned:
                 print("Warning: No JSON markers or clear brace match found. Using weak fallback.")
                 json_str_cleaned = generated_text.strip().split('\n')[0].strip()

            if not json_str_cleaned: raise ValueError("Could not extract any potential JSON content.")
            print(f"LLM Router: Attempting to parse JSON:\n{json_str_cleaned}")
            data = json.loads(json_str_cleaned)

            if "plot_type" not in data or "parameters" not in data: raise KeyError("Missing keys")
            params = data["parameters"]; required_params = ["city", "country", "start_year", "end_year"]
            if not all(k in params for k in required_params): raise KeyError(f"Missing params: Need {required_params}")
            try: start_year_int = int(params["start_year"]); end_year_int = int(params["end_year"])
            except ValueError as e: raise ValueError(f"Non-integer year: {e}.") from e

            request = PlotRequest(
                plot_type=PlotType(data["plot_type"]), city=str(params["city"]), country=str(params["country"]),
                start_year=start_year_int, end_year=end_year_int, title=params.get("title")
            )
            request.validate()
            print("LLM Router: Parsed and validated PlotRequest.")
            return request
        except (json.JSONDecodeError, KeyError, ValueError, TypeError) as e: print(f"LLM Router: Failed parsing/validating JSON: {e}. Raw output:\n{generated_text}"); return None
        except Exception as e: print(f"LLM Router: Unexpected parsing error: {e}"); return None

    def execute_request(self, request: PlotRequest) -> Optional[plt.Figure]:
        """Executes the data pipeline and plotting tool based on the request."""
        if not isinstance(request, PlotRequest): print("LLM Router: Invalid PlotRequest."); return None
        print(f"LLM Router: Executing request for {request.plot_type.name}..."); fig = None
        try:
            filtered = self.pipeline.filter_data(request.city, request.country, request.start_year, request.end_year)
            if filtered is None or filtered.empty: raise ValueError(f"No data found")
            yearly_avg = self.pipeline.calculate_yearly_avg(filtered)
            if yearly_avg is None or len(yearly_avg) < 2: raise ValueError(f"Not enough yearly avg data ({len(yearly_avg) if yearly_avg is not None else 0} points)")
            plot_func = self.registry.get_tool(request.plot_type.value)
            # --- ADD CHECK FOR plot_func ---
            if plot_func is None:
                raise ValueError(f"Plot function for '{request.plot_type.value}' not found in registry.")
            # ------------------------------
            title = request.title or f"Avg Temp {request.city}, {request.country} ({request.start_year}-{request.end_year})"
            print(f"LLM Router: Calling tool '{request.plot_type.value}'...")
            fig = plot_func(yearly_avg, title)
            if fig is None: raise RuntimeError("Plotting function returned None.")
            print("LLM Router: Plot generation successful.")
            return fig
        except (FileNotFoundError, ValueError, RuntimeError, KeyError) as e: print(f"LLM Router: Execution pipeline error: {e}"); return None
        except Exception as e: print(f"LLM Router: UNEXPECTED execution error: {e}"); import traceback; traceback.print_exc(); return None

# --- 13. Main Execution Logic ---
def main():
    print("--- Starting Main Execution (Structured Function Router v3) ---")
    # --- 1. Ensure Base Data Exists ---
    countries = ['Germany', 'Japan', 'France', 'United Kingdom', 'USA', 'Egypt', 'Australia', 'Russia', 'China', 'Italy', 'Spain']
    year_start = 1950; year_end = 2013; input_csv_path = TEMP_DATA_PATH
    print(f"\nEnsuring base data file exists at {DATA_CSV_PATH}...")
    setup_result = load_and_filter_data_for_setup(input_csv_path, countries, year_start, year_end)
    if setup_result is None: print("Exiting: Initial data setup failed."); return

    # --- 2. Generate Fine-tuning Data ---
    print("\n--- Generating Fine-tuning Data ---")
    function_call_df = None # Init
    try:
        df_for_training_gen = pd.read_csv(DATA_CSV_PATH); df_for_training_gen['dt'] = pd.to_datetime(df_for_training_gen['dt']); df_for_training_gen['year'] = df_for_training_gen['dt'].dt.year
        function_call_df = create_function_call_training_data(num_examples=5000) # Generate data
        del df_for_training_gen
    except Exception as e: print(f"Error preparing data for tuning generation: {e}"); return
    if function_call_df.empty: print("Exiting: Failed to generate training data."); return

    # --- 3. Fine-tune the Model ---
    print("\n--- Starting Fine-tuning Process ---")
    gc.collect(); torch.cuda.empty_cache()
    adapter_path = os.path.join(OUTPUT_DIR, "final_adapter")
    adapter_model_file = os.path.join(adapter_path, 'adapter_model.safetensors')
    if not os.path.exists(adapter_model_file):
        print(f"Adapter not found, starting fine-tuning...")
        adapter_path_result = finetune_phi2_qlora_structured(function_call_df) # Use the generated df
        if adapter_path_result is None: print("Exiting: Fine-tuning failed."); del function_call_df; gc.collect(); return
        adapter_path = adapter_path_result
    else: print(f"Found existing adapter at {adapter_path}. Skipping fine-tuning.")
    del function_call_df; gc.collect(); torch.cuda.empty_cache()
    print(f"Using adapter path: {adapter_path}")

    # --- 4. Initialize Components for Inference ---
    print("\n--- Initializing Components for Inference ---")
    router = None
    try:
        data_pipeline = DataPipeline(DATA_CSV_PATH)
        router = LLMFunctionRouter(MODEL_NAME, adapter_path, data_pipeline, registry)
        print("DataPipeline and LLMFunctionRouter initialized.")
    except Exception as e: print(f"Error initializing components: {e}"); return

    # --- 5. Test Inference & Execution ---
    print("\n" + "="*30 + " STARTING INFERENCE TESTS " + "="*30)
    test_queries = [
        f"Create a line plot showing the average temperature trends for Paris, France from 1960 to 2000?",
        f"Generate a scatter plot with its trend line for Tokyo, Japan temperature between 1970 and 2010.",
        f"Show me a line graph for Berlin, Germany, years 1980 to 2005.",
        f"Produce a scatter plot analysing the temperature trend in Sydney, Australia from 1955 to 1995.",
        f"Make a line plot for temperature in Tokyo, Japan for 1990-2010",
    #    f"I want a scatter graph with trend for Moscow, Russia, 1975 to 2005", # Should fail data pipeline
    #    f"Plot something random for Berlin 1999", # Should fail parsing
    #    f"Line plot for Berlin, Germany, 2015-2020" # Should fail data pipeline
    ]
    for query in test_queries:
        print(f"\n--- Processing Query: '{query}' ---")
        fig = None
        try:
            plot_request = router.parse_request(query)
            if plot_request:
                fig = router.execute_request(plot_request)
                if fig: print(f"â Success: Plot generated.")
                else: print(f"â Failed: Execution pipeline failed after parsing.")
            else: print(f"â Failed: Could not parse query into a valid request.")
            if fig: plt.show() # Display plot
        except Exception as e: print(f"â Error processing query loop: {e}")
        finally:
            if fig: plt.close(fig); print("Closed plot figure.")
            else: plt.close('all'); print("Closed all plot figures.")
            # Optional: Unload model if needed for memory between tests
            # router._unload_model()
            gc.collect(); torch.cuda.empty_cache()
        print("\n" + "="*70 + "\n")
    print("--- Main Execution Finished ---")

# --- Script Entry Point ---
if __name__ == "__main__":
    tokenizer_for_formatting = None
    main()
````
